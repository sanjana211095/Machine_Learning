```python
start = time.time()
alst = list(range(1,999999))
newlst = []
newlst = [e*e for e in alst]
end =time.time()
print("Time Took:" ,end-start)
```

```python
start1 =time.time()
alst = list(range(1,999999))
newlst = []
for elem in alst:
    newlst.append(elem*elem)
end1 =time.time()
print("Time Took:" ,end1-start1)
```

```python
start2 =time.time()
alst = list(range(1,999999))
newlst = []
newlst = list(map(lambda x : x*x, alst))
end2 =time.time()
print("Time Took:" ,end2-start2)
```

# Task 6

```python
data=[['t1',10,20],['t2',10,10],['t3',15,20],['t4',12,12],['t5',30,30],['t6',32,32],['t7',20,20],['t8','',''],['t9',10,""],['t10',11,11],
      ['t11','hai',20],['t12','?',30],['t13',0,0],['t13',0,0],['t15',0,'?']]
```

```python
task=pd.DataFrame(data,columns=['Names','Got','Exp'])
task
```

```python
task["Got"]=task["Got"].replace("hai","")
```

```python
task["Exp"]=task["Exp"].replace("None","")
```

```python
task["Exp"]=task["Exp"].replace("?","")
```

```python
task["Got"]=task["Got"].replace("?","")
```

```python
task=task.replace("",np.NaN)
```

```python
task
```

```python
A=task["Got"].median()
B=task["Exp"].median()
```

```python
A
```

```python
task["Got"]=task["Got"].fillna(A)
```

```python
task["Exp"]=task["Exp"].fillna(B)
task
```

```python
task["Got"]=task["Got"].astype(int)
```

```python
task["Exp"]=task["Exp"].astype(int)
task
```
