```python
df1=Data[["CO(GT)","PT08.S1(CO)","C6H6(GT)","PT08.S2(NMHC)",
          "NO2(GT)","PT08.S4(NO2)","PT08.S5(O3)","T","RH","AH"]]
sns.pairplot(df1)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
xvalues = np.array(Data["PT08.S2(NMHC)"]).reshape(-1,1)
yvalues = np.array(Data["PT08.S1(CO)"]).reshape(-1,1)

x_train,x_test,y_train,y_test = train_test_split(xvalues,yvalues,test_size= .20)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color = "b")
plt.plot(x_test, y_pred, color="r")
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test,y_pred)
r2
```
